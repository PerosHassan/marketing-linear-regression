# Simple Linear Regression - Marketing ROI Analysis
### 3MTT Evaluation Project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Load the cleaned dataset
df = pd.read_csv('marketing_and_sales_data_evaluate_lr.csv')
df = df.dropna()

print("--- Correlation Matrix ---")
print(df.corr(numeric_only=True)['Sales'])
```

### Model Implementation
Based on the data, TV marketing spend has the highest positive correlation with Sales. We will use it as our independent variable.

```python
X = df['TV']
y = df['Sales']
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

### Model Diagnostics & Interpretation
- **R-squared:** 0.99 (99% of variance in Sales is explained by TV marketing budget).
- **P-value:** < 0.05 (Highly statistically significant relationship).
- **Business Takeaway:** TV advertising yields the strongest ROI impact. Budget allocations should prioritize TV marketing to maximize sales volume.

```python
residuals = model.resid
fitted_values = model.fittedvalues

plt.figure(figsize=(6, 4))
sns.scatterplot(x=fitted_values, y=residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs Fitted Values')
plt.show()
```
